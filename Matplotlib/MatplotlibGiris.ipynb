```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # her seferinde plt.show() yazmamk için yazıldı. Notebook dışındaki editörlerde çalışmaz
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b") #renkler için "b" yazıldı
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
# plt.show() satır 4'e bak
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3 
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b.-") # "b--" de olabilirdi
```

```python
plt.subplot(1,2,1) #1 sıra olacak 2 kolon olacak şu an 1. grafiği çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2,"b.-")
plt.subplot(1,2,2) #1 sıra olacak 2 kolon olacak şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"b")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) ## ilk değerler nerede başlayacağını belirler-soldan ve aşağıdan- son iki değer ise genişlik ve yükseklik
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"b")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1 ,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"b")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()    
#benimFigure yazılması gerekebilir başka bir editör kullanılıyorsa    
```

```python
yeniFİgure = plt.figure()

yeniEksen = yeniFİgure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3 , label = "numpyDizisi ** 3")
yeniEksen.legend(loc=6)
```

```python
yeniFİgure = plt.figure(figsize=(6,6)) # dpi=100

yeniEksen = yeniFİgure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3 , label = "numpyDizisi ** 3")
yeniEksen.legend(loc=6)
```
