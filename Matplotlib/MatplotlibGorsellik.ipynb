```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2 
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"#33C9C9")
benimEksen.plot(numpyDizisi2,numpyDizisi1,"#C93395",alpha=0.4)
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="#C93395",linewidth=4.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3 ,color= "b",linewidth=7.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4 ,color="#496210",linestyle="-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5 ,color="#496210",linestyle=":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6 ,color="#496210",linestyle="--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7 ,color="#1B042A",linestyle="--",marker="o",markersize=4,markerfacecolor="r")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8 ,color="#1B042A",linestyle="--",marker="+",markersize=4,)
```
